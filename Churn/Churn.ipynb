## Churn Predictions

```python
# required imports
import pandas as pd
from sklearn.model_selection import train_test_split
import os
```

```python
df = pd.read_csv('./data/churn.csv')
df.head(5)
```

```python
# convert categorical data into dummy data
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis = 1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
# Split train test data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(f"Train : {X_train.shape}, Test : {y_train.shape}")
```

```python
X_train.head()
```

```python
y_train.head()
```

## Importing dependency


```python
# Sequential is base class of tensorflow models
# layers are the types how you want to build the model. LSTm for NLP and Conv2D for image
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential() # instantiating model class sequential
# adding hidden layers
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns))) # 32 neurons
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid')) # output will be of type sigmoid and only one value/output
```

## Build and compile model

```python
# compiling model we tell tf how to build the model with its features like sgd (random selector)
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

## Fit, Predict and Evaluate model

```python
model.fit(X_train,y_train, epochs=300, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if x < 0.5 else 1 for x in y_hat]
```

```python
score = accuracy_score(y_test, y_hat)
print(f"Current model accuracy is {round(float(score) * 100, 2)}%")
```

## Saving & Reloading model

```python
model.save('./data/churntfmodel')
```
